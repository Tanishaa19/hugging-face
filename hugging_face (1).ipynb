```python
from transformers import pipeline

summarizer = pipeline("summarization",model="facebook/bart-large-cnn")

text = """
Artificial Intelligence is transforming industries by automating repetitive tasks,
improving efficiency, and helping organizations make better decisions using data.
AI is widely used in healthcare, education, finance, cybersecurity, and many
other sectors across the world.
"""

summary = summarizer(text,max_length=60,min_length=20)

print("\nSUMMARY \n")
print(summary[0]['summary_text'])
```

```python
translator = pipeline("translation_en_to_hi",model="Helsinki-NLP/opus-mt-en-hi")

def translate_text(text):
    result = translator(text)
    return result[0]['translation_text']

text = "Artificial Intelligence is changing the world."

print("TRANSLATION:\n", translate_text(text))
```

```python
!pip install gtts -q

from gtts import gTTS
from IPython.display import Audio
from google.colab import files


translation = "Artificial Intelligence is changing the world."

tts = gTTS(
    text=translation,
    lang='en'
)

audio_file = "translated_audio.mp3"

tts.save(audio_file)

print("Audio file saved successfully!")

Audio(audio_file)

files.download(audio_file)
```
